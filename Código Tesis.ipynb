```python
#Ajuste lineal robusto
def bisquare_weights(residuals, scale):
    weights = np.zeros_like(residuals)
    for i, r in enumerate(residuals / scale):
        if np.abs(r) < 1:
            weights[i] = (1 - r**2)**2
        else:
            weights[i] = 0
    return weights

def robust_scale(residuals):
    return 1.4826 * np.median(np.abs(residuals - np.median(residuals)))

def robust_linear_fit(X, y, max_iterations=50, tol=1e-6):
    n = len(y)
    X = np.column_stack((np.ones(n), X))
    beta = np.linalg.lstsq(X, y, rcond=None)[0]
    
    for _ in range(max_iterations):
        residuals = y - X @ beta
        scale = robust_scale(residuals)
        weights = bisquare_weights(residuals, scale)
        
        W = np.diag(weights)
        beta_new = np.linalg.lstsq(X.T @ W @ X, X.T @ W @ y, rcond=None)[0]
        
        if np.linalg.norm(beta_new - beta) < tol:
            break
        beta = beta_new

    return beta, weights

np.random.seed(42)
```
